### query() rules
- Expression as a string
- Column based reference
- Backticks for col_name with space or special chars
- Operators (&, |,~,>=,>,<=,<,== etc,)
- @ to referance Python vars
- Chained comparison

```python
# Filtering of Data
df[ df["aqi"] > 100]
df[ (df["aqi"] > 100) & (df["temperature"] > 30)]
df[ (df["aqi"] > 100) & (df["temperature"] < 30)]

aqi_data = df[ (df["aqi"] > 100) & (df["temperature"] > 30)][["city", "aqi"]]

aqi_data
aqi_data.iloc[0]
aqi_data.loc[6]

# Query
df.query("aqi > 100")
df.query("aqi > 100 & temperature > 30")[["city", "aqi"]]
```

```python
# Filtering of Data
df[ df["aqi"] > 100]
df[ (df["aqi"] > 100) & (df["temperature"] > 30)]
df[ (df["aqi"] > 100) & (df["temperature"] < 30)]

aqi_data = df[ (df["aqi"] > 100) & (df["temperature"] > 30)][["city", "aqi"]]

aqi_data
aqi_data.iloc[0]
aqi_data.loc[6]

# Query
aqi_val = 100
df.query("aqi > 100")
df.query("aqi > @aqi_val & temperature > 30")[["city", "aqi"]]
```

```python
# Filtering of Data
df[ df["aqi"] > 100]
df[ (df["aqi"] > 100) & (df["temperature"] > 30)]
df[ (df["aqi"] > 100) & (df["temperature"] < 30)]

aqi_data = df[ (df["aqi"] > 100) & (df["temperature"] > 30)][["city", "aqi"]]

aqi_data
aqi_data.iloc[0]
aqi_data.loc[6]

# Query - copy
aqi_val = 100
df.query("aqi > 100")
aqi_data = df.query("aqi > @aqi_val & temperature > 30")[["city", "aqi"]].copy()
```
